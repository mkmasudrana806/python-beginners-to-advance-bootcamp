# Problems with working in text mode
* can't work with binary files like images, audio and video
* not good for other data types like int/float/list/tuples



```python
# working with binary file. it gives error if we want to read with 'r' mode only
with open('./files/image.png', 'r') as f:
    s = f.read()
    print(s)
```

```python
# Note: we can't read any bianry data. 
# working with binary file with mode 'rb' 
with open('./files/image.png', 'rb') as f:
    s = f.read()
    print(s)
```

```python
# copy binary file to another file. we can save binary file to anywhere
# but it is not efficient way to work with binary data
with open('./files/image.png', 'rb') as f:
    with open("./files/copy_image.png", "wb") as wf:
        wf.write(f.read())
```

# working with other type data
int / list / dictionary / tuple

```python
# 
# int not possible
with open('./files/sample3.txt', 'w') as f:
    f.write(10) # write() argument must be str, not int
    
    
# int possible ( you need to convert int to string)
with open('./files/sample3.txt', 'w') as f:
    f.write("10") # write() argument must be str, not int
```

```python
# dictionary not possible
person = {
    'name': "Masud rana", 
    'age': 23, 
}
with open('./files/sample3.txt', 'w') as f:
    f.write(person) # write() argument must be str, not int


# dictionary possible ( need to convert dict to string)
person = {
    'name': "Masud rana", 
    'age': 23, 
}
with open('./files/sample3.txt', 'w') as f:
    f.write(str(person)) # write() argument must be str, not int
```
