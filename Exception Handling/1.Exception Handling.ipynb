# Exceptions Error
if things go wrong during the executing of the program(runtine). it generally happens when something unforeseen has happened.
* Exceptions are raised by python runtime
* You have to takle is on the fly

## Examples
* Memory overflow
* Divisible by 0 -> logical error
* Database error

```python
# Divide by zero (0) error
def divide_num(num):
  r = 12 / num
  return r

print(divide_num(0))
```

```python
# memory overflow data
def print_name(name):
  print('Hello ', name)
  print('How are you?')
  print('What are you doing?')
  print_name(name)

print_name('Masud Rana')
```

# Exception Handling using try..catch

```python
# let's read a file and handle exception
# file.txt doesn't exist. it simply stop the program.
# but due to exception handling, the program will not stop program
try:
  with open("file.txt", "r") as f: # file.txt file doesn't exists. it throw error but program will not stop
    print(f.read())
except:
  print("File not found")
```
