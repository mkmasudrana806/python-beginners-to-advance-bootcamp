# Nested Functions
function inside a function is called nested functions

```python
# nested function example
# Note: local function can not be called from outside of the function
def f():
  def g():
    print('inside function g')
  print('insdie function f')

f() # what will be the output?
# g() # errors. local function g() can not ba called outsite of the local scope

# nested function
def f():
  def g():
    print('inside function g')
  g()
  print('insdie function f')

f() # what will be the output?



# recursion or stack error
def f():
  def g():
    print('inside function g')
  g()
  f() # recursively it will call f() infinite times
  print('insdie function f')

f() # what will be the output?
```
